#### **Age Group Segmentation**

<ul>
    <li>Young Adults: 18 - 24 years</li>
    <li>Young Professionals: 25 - 34 years</li>
    <li>Established Adults: 35 - 44 years</li>
    <li>Middle-Aged Consumers: 45 - 54 years</li>
    <li>Seniors: 55 - 65 years</li>
</ul>

```python
#Addition of new column - 'Age Group'
def assign_age_group(age):
    if age < 18:
        return "Minors"
    elif 18 <= age <= 24:
        return "Young Adults"
    elif 25 <= age <= 34:
        return "Young Professionals"
    elif 35 <= age <= 44:
        return "Established Adults"
    elif 45 <= age <= 54:
        return "Middle-Aged Consumers"
    else:
        return "Seniors"

df['age_group'] = df['age'].apply(assign_age_group)
cols = list(df.columns)
df = df[cols[:-2] + [cols[-1], cols[-2]]]
df.head(10)
```

## Data Preparation

```python
X = df.drop(columns = ["churn", "age_group"])
y = df["churn"]

X.shape
y.shape
```

```python
from sklearn.model_selection import train_test_split

#Dataset Splitting: Train Set (70%), Validation set (30%)
X_train, X_val, y_train, y_val = train_test_split(X, y, stratify = y, test_size = 0.3, random_state = 0)

#Structure of Data Split
print("X_train shape:", X_train.shape)
print("X_val shape:", X_val.shape)
```

```python
#Preprocessing: Scaling the Categorical and Numerical Columns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer

col_num = ["age", "tenure", "usage_frequency",	"support_calls", "payment_delay", "total_spend", "last_interaction"]
col_cat = ["gender", "subscription_type", "contract_length"]

preprocessor = make_column_transformer((OneHotEncoder(handle_unknown = "ignore"),col_cat),remainder='passthrough')
```

## Model Development

### Random Forest on Train Set - no hyperparameters

```python
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV


# Define pipeline with preprocessing, feature selection, and classifier
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("feature_selection", VarianceThreshold(threshold=0.01)),
    ("randomforestclassifier", RandomForestClassifier(random_state=0))  
])

rf1 = pipeline.fit(X_train, y_train)
```

### Evaluation on Train & Validation Set

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Predict on the val set
y_pred = rf1.predict(X_val)
y_pred_proba = rf1.predict_proba(X_val)[:, 1]  # Get probabilities for ROC-AUC score

# Compute evaluation metrics
conf_matrix = confusion_matrix(y_val, y_pred)
accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)
roc_auc = roc_auc_score(y_val, y_pred_proba)

# Display results
metrics_results = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1,
    "ROC-AUC Score": roc_auc
}

metrics_df = pd.DataFrame(metrics_results, index=["Score"])
print(metrics_df)

# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
import seaborn as sns


# Plot frequency distribution
plt.figure(figsize=(8, 6))
sns.histplot(y_pred_proba, bins=50, color='purple', alpha=0.7)

# Add labels and title
plt.xlabel("Predicted Probability")
plt.ylabel("Frequency")  # Shows count instead of density
plt.title("Frequency Distribution of Predicted Probabilities")
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC-AUC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line (random chance)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.show()
```

## Problems:

- The model is overfitting, since this highly unusual in real world scenarios
- There is slight class imbalance
- Also, we need to adjust the classification threshold to reduce false negatives and increase false positives, since model is misclassifying churn as not churn. For our business problem, it is okay if the model classifies not churned as churned. So, we need to increase the TPR and FPR. 

## Random Forest Classifier - with hyperparameter tuning

```python
from sklearn.model_selection import StratifiedKFold

# Define pipeline with preprocessing, feature selection, and classifier
pipeline = Pipeline([
    ("preprocessor", preprocessor),  # Step 1: Encode categorical features
    ("feature_selection", VarianceThreshold(threshold=0.01)),  # Step 3: Feature selection
    ("classifier", RandomForestClassifier(random_state=0))  # Step 4: Train classifier
])

#Stratified K-fold
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

# Define hyperparameter grid for tuning
rf_param_grid = {
    "classifier__n_estimators": [10, 25],  
    "classifier__max_depth": [1, 2, 3],  
    "classifier__max_features": ["sqrt"], 
    "classifier__min_samples_split": [3, 5],  
    "classifier__min_samples_leaf": [3, 4],  
    "classifier__max_samples": [0.5, 0.75] 
}

#Grid Search CV
rf_grid = GridSearchCV(
    pipeline, param_grid=rf_param_grid, cv=cv_strategy, n_jobs=-1, scoring="accuracy", return_train_score=True
)

# Fit the model
model1 = rf_grid.fit(X, y)

# Output best parameters
print("Best parameters:", rf_grid.best_params_)
print("Best cross-validation score:", rf_grid.best_score_)
```

```python
# Predict on the val set
y_pred = model1.predict(X)
y_pred_proba = model1.predict_proba(X)[:, 1]  # Get probabilities for ROC-AUC score

# Compute evaluation metrics
conf_matrix = confusion_matrix(y, y_pred)
accuracy = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
f1 = f1_score(y, y_pred)
roc_auc = roc_auc_score(y, y_pred_proba)

# Display results
metrics_results = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1,
    "ROC-AUC Score": roc_auc
}

metrics_df = pd.DataFrame(metrics_results, index=["Score"])
print(metrics_df)

# Plot confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
# Compute ROC curve
fpr, tpr, thresholds = roc_curve(y, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot ROC-AUC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line (random chance)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.show()
```
